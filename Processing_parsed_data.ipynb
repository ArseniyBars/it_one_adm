```python
new_subregions=['Новомосковский административный округ', 'Троицкий административный округ', 
                'Зеленоградский административный округ', 'Зеленоград', 'район Троицк']
def expand_adress(r):
    res={'precise_adress': None, 'house': None, 'street': None, 'quartal': None, 'city_distinct': None, 
         'city': None, 'distinct': None, 'subregion': None, 'region': None, 'post_index': None}
    adress=[s.strip() for s in r['adress'].split(',')]
    if adress[-1].isdigit():
        res['post_index']=adress[-1]
        adress=adress[:-1] 
        if adress==[]:
            return pd.concat([r, pd.Series(res)])
    else:
        post_idex=None
    if adress[-1]=='Москва' or adress[-1].endswith('область'):
        res['region']=adress[-1]
        adress=adress[:-1]
        if adress==[]:
            return pd.concat([r, pd.Series(res)])
        if ('округ' in adress[-1] and res['region']=='Москва') or adress[-1] in new_subregions[3:]:
            res['subregion']=adress[-1]
            if adress[-1]=='район Троицк':
                res['distinct']='район Троицк'
                res['city']='Троицк'
            adress=adress[:-1]
            if adress==[]:
                return pd.concat([r, pd.Series(res)])
    if res['region']=='Москва' and res['subregion'] not in new_subregions:
        if adress[-1]=='Троицк':
            res['city']='Троицк'
            res['distinct']='район Троицк'
            adress=adress[:-1]
            if adress==[]:
                return pd.concat([r, pd.Series(res)])
        else:
            res['city']='Москва'
    elif res['subregion'] in ['Зеленоград', 'Зеленоградский административный округ']:
        res['distinct']=res['subregion']
        res['city']='Зеленоград'
    else:
        if 'округ' in adress[-1] or 'район' in adress[-1]:
            res['distinct']=adress[-1]
            adress=adress[:-1]
            if adress==[]:
                return pd.concat([r, pd.Series(res)])
        if adress[-1][0].isdigit():
            if adress[0].split(' ')[0] in ('деревня', 'посёлок', 'село'):
                res['city']=adress[0]
        else:
            if adress[-1].startswith('квартал') or adress[-1].startswith('жилой комплекс'):
                if res['distinct'] is None:
                    raise Exception('Incorrect adress')
                res['city']=re.sub(r'^район ', '', res['distinct'])
            else:
                res['city']=adress[-1]
                if res['distinct'] is None:
                    res['distinct']=res['city']
                adress=adress[:-1]
                if adress==[]:
                    return pd.concat([r, pd.Series(res)])              
    if adress[-1].split(' ')[0] in ('район', 'микрорайон', 'городок', 'посёлок') \
                    or re.fullmatch(r'(\d+)-й микрорайон', adress[-1]):
        res['city_distinct']=adress[-1]
        adress=adress[:-1]
        if adress==[]:
            return pd.concat([r, pd.Series(res)])
    if re.match(r'(\d+-й )?(микрорайон)|(квартал)', adress[-1]) or adress[-1].startswith('жилой комплекс'):
        res['quartal']=adress[:-1]
        adress=adress[:-1]
        if adress==[]:
            return pd.concat([r, pd.Series(res)])
    res['precise_adress']=', '.join(adress)
    if 'подъезд' in adress[-1]:
        adress=adress[:-1]
        if adress==[]:
            return pd.concat([r, pd.Series(res)])
    if re.fullmatch(r'((вл)|(д))?(\d+)[А-Я]?([/]\d+)?(к\d+)?(с\d+)?', adress[-1]):
        res['house']=adress[-1]
        if len(adress)>1:
            res['street']=adress[-2]
    return pd.concat([r, pd.Series(res)])
df_parsed1=df_parsed.apply(expand_adress, axis=1, result_type='expand')
```

```python
metro_stations_center=["Киевская", "Краснопресненская", "Белорусская", "Новослободская", "Проспект Мира",
                       "Комсомольская", "Курская", "Таганская", "Павелецкая", "Добрынинская", "Октябрьская",
                       "Парк культуры", "Комсомольская", "Красные Ворота", "Чистые пруды", "Лубянка",
                       "Охотный Ряд", "Библиотека имени Ленина", "Кропоткинская", "Сухаревская",
                       "Тургеневская", "Китай-город", "Третьяковская", "Менделеевская", "Цветной бульвар",
                       "Чеховская", "Боровицкая", "Полянка", "Серпуховская", "Маяковская", "Тверская",
                       "Театральная", "Новокузнецкая", "Баррикадная", "Пушкинская", "Кузнецкий мост",
                       "Смоленская", "Арбатская", "Площадь Революции"]
df_cleaned['is_center_of_Moscow']=df_cleaned.metro_station.isin(metro_stations_center
                                                               ).fillna(False).astype(int)
df_cleaned[(df_cleaned.region=='Москва')&(~df_cleaned.subregion.isin(new_subregions))&
    (df_cleaned.is_center_of_Moscow==0)].join(
    pd.read_csv('target_nov_sept.csv').set_index('TID'))['target_sept'].to_csv(
        'target_september_selected.csv')
```

```python
df_cleaned['is_Moscow']=df_cleaned.metro_station.notnull()|(
    (df_cleaned.region=='Москва')&(~df_cleaned.subregion.isin(new_subregions)))
non_moscow=df_cleaned[~df_cleaned.is_Moscow][['latitude', 'longitude', 'distinct', 'city', 'distance_to_mkad']]
non_moscow=non_moscow[non_moscow.city.notnull().copy()]
distincts=pd.read_csv('distincts.csv').set_index(['distinct'])
def distance_to_distinct_center(inp):
    dc=distincts.loc[inp.distinct]
    return geodesic((inp.latitude, inp.longitude), (dc.latitude, dc.longitude)).km
non_moscow['distance_to_distinct_center']=non_moscow.apply(distance_to_distinct_center, axis=1)
populations=pd.read_csv('populations.csv').set_index(['distinct', 'city']).population
non_moscow['is_rural']=(non_moscow.city.str.startswith('деревня')|
                        non_moscow.city.str.startswith('село')).astype(int)
non_moscow['population']=non_moscow.apply(lambda r: populations[r['distinct'], r['city']], axis=1)
non_moscow[['distance_to_mkad', 'distance_to_distinct_center', 'is_rural', 'population']
        ].to_csv('non_moscow.csv')
```

```python
new_metro_stations=['Университет дружбы народов', 'Генерала Тюленева', 'Тютчевская', 'Корниловская',
                    'Потапово', 'Пыхтино', 'Аэропорт Внуково', 'Кленовый бульвар', 'Нагатинский Затон',
                    'Физтех', 'Лианозово', 'Яхромская']
moscow_data=df_cleaned[df_cleaned.is_Moscow&(~df_cleaned.metro_station.isin(new_metro_stations))][
                        ['latitude', 'longitude', 'is_center_of_Moscow', 'distance_to_station']].copy()
moscow_data['is_center']=moscow_data.is_center_of_Moscow.apply(int)

def distance_to_metro_station(inp):
    if pd.isnull(inp):
        return None
    dist, un=inp.split()[:2]
    if un=='м':
        return int(dist)
    elif un=='км':
        dist=dist.replace(',', '.')
        return float(dist)*1000
    else:
        raise Exception(f'Unrecognizable: {inp}')
moscow_data['distance_to_metro_station']=moscow_data.distance_to_station.apply(distance_to_metro_station)

def distance_to_center(inp):
    return geodesic((inp.latitude, inp.longitude), (55.755864, 37.617698)).km
moscow_data['distance_to_center']=moscow_data.apply(distance_to_center, axis=1)

moscow_data[['latitude', 'longitude', 'is_center', 'distance_to_center', 'distance_to_metro_station']].to_csv('moscow_data.csv')
```
